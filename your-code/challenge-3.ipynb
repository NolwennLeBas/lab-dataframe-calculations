## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
#step1 : we need to group rows by customer_id and calculate the sum of amount_spent
orders2=orders[['CustomerID','Country', 'amount_spent']]
orders2=orders2.groupby(['CustomerID', 'Country'], as_index=False)['amount_spent'].sum()
orders2
```

```python
#two groups of customers. 
#VIP Customers above the 95th percentile (aka. 0.95 quantile). 
#Preferred Customers between the 75th and 95th percentile.

#step2: calculating the percentiles
p_95 = np.percentile(orders2, 95)
p_75 = np.percentile(orders2, 75)
display(p_95, p_75)
```

```python
#step3: creating a function to evaluate the category of amount_spent for each customerID
def quantile_range(row):
    if row['amount_spent'] >= p_95:
        return 'VIP'
    elif row['amount_spent'] < p_95 and row['amount_spent'] >= p_75:
        return 'Preferred'
    else:
        return ''
        
#step4: applying the function and assigning the result to a new column, labelling appropriately each Customer
orders2['Label']=orders2.apply(quantile_range, axis=1)
orders2
```
